<h3>Theory</h3>
<p>If the key is closed in the circuit of fig 1, the dc emf soucre will charge the capacitor through the resistance. Then charge in the capacitor at any instance is given as: </p>


\begin{align*}
Q = Q'(1-e^{\frac{-1}{RC}}) &...................     (1)\\
\end{align*}

<p>Where Q' is the maximum charge to be stored in the capacitor, R and C are the resistance of the capacitor in the circuit. The diffrentiating (1), charging current is obtained as: </p>

\begin{align*}
I = I'(e^{\frac{-1}{RC}}) &...................     (2)\\
\end{align*}

Now, if the source is disconnected and points A and B are connected, the capacitor will discharge through the resistor. Then charge in the capacitor at any instance during discharge is given as:

\begin{align*}
Q = Q'(e^{\frac{-1}{RC}}) &...................     (3)\\
\end{align*}

and discharging current is given by:
\begin{align*}
I = I'(e^{\frac{-1}{RC}}) &...................     (4)\\
\end{align*}

<img src="capacitor.jpg" width="300" height="200">

Note that the discharging current is in opposite direction of that of charging current. Here if a graph is drawn between 'I' versus 't' for the equations (2) and (4). exponentially decaying curves obtained. Then taking logarithm of equns (2) or (4),

\begin{align*}
ln(I) = ln(I') - \frac{-1}{RC} &...................     (5)\\
\end{align*}

If a graph is plotted between 'ln(I)' & 't', a straight line with slope $-\frac{1}{RC}$ is obtained


```python
#importig the libraries
import numpy as np
import matplotlib.pyplot as plt

capcitances = []

x = np.linspace(-5,5,100)
y = -2*x-5
plt.plot(x, y, '-b', label="I = I'-1/RC")
plt.xticks([])
plt.yticks([])
plt.title("Graph of I = I'-1/RC")
plt.xlabel('t', color='#1C2833')
plt.ylabel('ln(I)', color='#1C2833')
plt.legend(loc='upper right')
plt.show()
```

If a point where $I=\frac{I'}{2} \ (t = T\frac{1}{2})$ is considered, eq(5) becomes:
\begin{align*}
ln(\frac{1}{2}) = -\frac{T1/2}{RC}\\
or, RC = \frac{T1/2}{ln2}\\
or, C = \frac{T1/2}{Rln2}&......(6)\\
\end{align*}

Similarly, if slope of the <b>the straight</b> line in above graph,
\begin{align*}
m = \frac{a}{b} = \frac{1}{RC}\\
or, RC = \frac{1}{m}\\
or, C = \frac{1}{Rm}&......(7)\\
\end{align*}

<h4>For R = 10Kohm</h4>
<h3>Charging</h3>

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()  # create object for the class

chg1_data = pd.read_csv("charging1.csv", names=['time','current']);
chg1_data = chg1_data[1:]
chg1_data["ampere"] = pd.to_numeric(chg1_data["current"])*0.05*(1/1000)
chg1_data["time"] = pd.to_numeric(chg1_data["time"])
chg1_data["lnI"] = np.log(chg1_data["ampere"])
chg1_data
```

### Plotting I(current) vs T(time) graph

```python
#plotting the points
x = chg1_data["time"]
y = chg1_data["ampere"]
plt.ylim(0,0.00085)
plt.xticks(range(0, 65))
plt.scatter(x,y)
plt.title("I vs time")
plt.xlabel("Time(t)")
plt.ylabel("I")
plt.plot(x,y)
plt.show()

t = x[5]
R = 10*1000
print("Here value corresponding to I'/2 is "+ str(t))
print("C = T/Rln2")
C = t/(R*0.69)*10**6
capcitances.append(C)
print("C = "+str(C)+" micro farads")
capcitances.append(C)
```

```python
#plotting the points
#TODO://Using the linear regression to find the straight line for the given points

x = chg1_data["time"].values.reshape(-1,1)
y = chg1_data["lnI"].values.reshape(-1,1)


linear_regressor.fit(x, y)  # perform linear regression
Y_pred = linear_regressor.predict(x)  # make predictions
m = linear_regressor.coef_[0]
plt.scatter(x,y)
plt.title("ln(I) vs time")
plt.xlabel("Time(t)")
plt.ylabel("ln(I)")
plt.plot(x, Y_pred, color='green')
plt.legend(("Regression Line","Joing the points"))
plt.show()

print("From the graph slope was: "+str(m[0]))
print("We have resistance R = 10k ohm")
print("Then, C = 1/Rm")
C = abs((1/(10**4 * m[0]))*10**6)
capcitances.append(C)
print("C = "+ str(C)+ " micro farad")
```
